```python
import numpy as np
from flair.embeddings import WordEmbeddings
from flair.embeddings import TransformerWordEmbeddings
from flair.data import Sentence
from scipy.spatial import distance
```

# **Glove Embeddings Example**

```python
glove_embedding = WordEmbeddings('glove')
#We know that Word2Vec model creates vector for a word that is independent of the context. 
#ie, Numeric representation of a word will remain same irrespective of the context of a particular word.
#Glove is also a WordEmbedding model
```

```python
sentence_1 = Sentence("The founder of Apple is Steve Jobs")
```

```python
glove_embedding.embed(sentence_1)
```

```python
for token in sentence_1:
  print(token)
  print(token.embedding)
  print("\n")
```

# **Glove Distance between the same word**

```python
#Wkt, in sent_1 --> Apple is pointing towards an organization.
#And in sent_2 --> It is pointing towards an apple.

glove_dist = distance.euclidean(np.array(sentence_1[3].embedding),
                                np.array(sentence_2[1].embedding))
print("Distance between Apple embeddings for Glove = {}".format(glove_dist))

#Note: In GPU This script will never work

#TAKEAWAY : So Word2Vec Model dont understand the context of the word. To prove that hypothesis, we have taken the vectors representations of Apple in both the sentences.
#Lets calculate the euclidean distance of the word Apple in two sentences. 
#Zero difference. So Word2Vec is context independent. ie, never understand the context.
```
